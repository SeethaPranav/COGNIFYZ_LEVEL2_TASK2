## Identify the most common combinations of cuisines

### 1.Explore the 'Cuisines' column

```python
print(data['Cuisines'].head())
```

### 2. Clean and count combinations

```python
# Drop rows with missing cuisines
data_cuisine = data.dropna(subset=['Cuisines'])

# Count top combinations
cuisine_combo_counts = data_cuisine['Cuisines'].value_counts().head(10)

print("Top 10 most common cuisine combinations:")
print(cuisine_combo_counts)
```

## Determine if certain combinations tend to have higher ratings.

```python
# Group by cuisine combination and get average rating
cuisine_rating = data_cuisine.groupby('Cuisines')['Aggregate rating'].mean().round(2)

# Combine count and rating for top combinations
top_cuisine_combos = pd.DataFrame({
    'Count': cuisine_combo_counts,
    'Average Rating': cuisine_rating[cuisine_combo_counts.index]
}).sort_values(by='Count', ascending=False)

print("Top cuisine combinations with their average ratings:")
print(top_cuisine_combos)
```

## Bar Chart of Ratings for Top Cuisine Combinations

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top_cuisine_combos.reset_index(), 
            x='Average Rating', y='Cuisines', palette='coolwarm')

plt.title('Average Ratings of Top Cuisine Combinations')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine Combinations')
plt.xlim(0, 5)
plt.grid(True)
plt.show()
```
